# Inference from the analysis
All the variables show a positive skewness, except for Education.

Age, Distance from home, Education and Percent salary hike show a negative kurtosis, so they are platyokurtic.

Monthly income, Number of companies worked, Total working years, Training times last year, Years at company, Years since last promotion and Years with current manager show a positive kurtosis, so they are Leptokurtic.

The Mean Monthly Income’s interquartile range (IQR) is at 54,680 suggesting company wide attrition across all income bands.

Mean age’s interquartile range (IQR) is at 13 forming a near normal distribution.

```python
from scipy.stats import pearsonr
```

```python
stats, p = pearsonr(df.Attrition, df.DistanceFromHome)
print("The r value is: ", stats, "\nThe p value is: ", p)

if(p < 0.05):
    print("The null hypothesis is rejected, so there is significant correlation between both variables.")
else:
    print("The null hypothesis is accepted, so there is no significant correlation between both variables.")
```
