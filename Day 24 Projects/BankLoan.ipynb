```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx", sheet_name=1)
```

```python
data.columns
```

```python
dataset = data.drop(['ID', 'ZIP Code'],axis=1)
```

```python
dataset.dropna()
```

```python
dataset.drop_duplicates()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
dataset['CCAvg']=np.round(dataset['CCAvg'])
```

```python
rf = RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
features = ['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education', 'Mortgage', 'Securities Account','CD Account', 'Online', 'CreditCard']
```

```python
rf.fit(X=dataset[features],y=dataset['Personal Loan'])
```

```python
print('OOB Accuarcy:')
print(rf.oob_score_);
```

```python
for feature, imp in zip(features, rf.feature_importances_):
    print(feature,imp);
```

```python
from sklearn import tree
```

```python
tree_model = tree.DecisionTreeClassifier()
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth=6, max_leaf_nodes=10)
```

```python
predictors = pd.DataFrame([dataset['Education'],dataset['CCAvg'], dataset['Income']]).T
```

```python
tree_model.fit(X=predictors, y=dataset['Personal Loan'])
```

```python
with open("Dtree.dot", 'w') as f:
    f = tree.export_graphviz(tree_model, feature_names=['Education', 'CCAvg', 'Income'], out_file=f);
```
