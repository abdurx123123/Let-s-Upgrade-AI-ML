```python
import pandas as pd
```

```python
dataset = pd.read_csv("general_data.csv")
dataset.head()
```

```python
from scipy.stats import wilcoxon
stats,p=wilcoxon(dataset.TotalWorkingYears,dataset.TrainingTimesLastYear)
print(stats,p)
if p < 0.05:
    print("Reject Null HYPOTHESIS")
else:
    print("Accept Null HYPOTHESIS")
```

```python
from scipy.stats import wilcoxon
stats,p=wilcoxon(dataset.Age,dataset.DistanceFromHome)
print(stats,p)
```

```python
from scipy.stats import kruskal
stats,p=kruskal(dataset.Attrition,dataset.Department,dataset.EducationField)
print(stats,p)
```
